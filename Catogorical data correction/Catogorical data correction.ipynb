```python
columns_to_drop = ['Unique ID', 'Annual Family Income ($)', 'IQ', 'Migrated within country',
                   'Maritial Status (0 - Single, 1 - Married, 2 - Divorced)', 'Education Status',
                   'Number of friends', 'Wants to change career', 'Has debt']

df_cleaned = df.drop(columns=columns_to_drop)

```

```python
print(df_cleaned.columns.tolist())
```

```python
df_cleaned.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Columns to label encode
columns_to_label_encode = [
    'Gender', 'English speaker', 'Migrated overseas', 'Has Diabetes', 'Has Gym Subscription',
    'Has OTT subscription', 'Likes spicy food', 'Likes desserts', 'Has kids', 'Drinks alcohol',
    'Smoker', 'Wear Specs', 'Education Status_encoded', 'Gender_encoded', 'English speaker_encoded'
]

# Perform label encoding for each column
for col in columns_to_label_encode:
    df_cleaned[col + '_encoded'] = label_encoder.fit_transform(df_cleaned[col])

# Drop the original columns after encoding
df_cleaned.drop(columns=columns_to_label_encode, inplace=True)

```

```python
# Print the label encoded DataFrame as an array
print(df_cleaned.to_numpy())

```

```python
# Perform One-Hot Encoding for 'Country' and 'Education Status'
one_hot_encoded_df = pd.get_dummies(df_cleaned)

# Print the one-hot encoded DataFrame as an array
print(one_hot_encoded_df.to_numpy())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar Chart for 'Country' Distribution
country_columns = ['Country_CAN', 'Country_IND', 'Country_JPN', 'Country_KOR']
country_counts = df_cleaned[country_columns].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title('Distribution of Clients Across Countries')
plt.xlabel('Country')
plt.ylabel('Number of Clients')
plt.show()

# Histogram for 'Age' Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Age'], bins=20, kde=True)
plt.title('Distribution of Age Among Clients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Pie Chart for 'Gender' Distribution (if available)
if 'Gender_encoded_encoded' in df_cleaned.columns:
    gender_counts = df_cleaned['Gender_encoded_encoded'].value_counts()
    plt.figure(figsize=(8, 8))
    plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Gender Distribution Among Clients')
    plt.show()

# Bar Chart for Binary Variables
binary_variables = ['Has Diabetes_encoded', 'Has Gym Subscription_encoded', 'Has OTT subscription_encoded',
                    'Likes spicy food_encoded', 'Likes desserts_encoded', 'Has kids_encoded',
                    'Drinks alcohol_encoded', 'Smoker_encoded', 'Wear Specs_encoded']
plt.figure(figsize=(12, 8))
for var in binary_variables:
    counts = df_cleaned[var].value_counts()
    sns.barplot(x=counts.index, y=counts.values, alpha=0.7, label=var.replace('_encoded', '').replace('_', ' '))
plt.title('Proportion of Clients with Binary Attributes')
plt.xlabel('Attribute')
plt.ylabel('Number of Clients')
plt.legend()
plt.show()

# Count Plot for Encoded Variables
encoded_variables = ['Migrated overseas_encoded', 'Education Status_encoded_encoded', 
                     'English speaker_encoded_encoded']
plt.figure(figsize=(12, 8))
for var in encoded_variables:
    sns.countplot(x=var, data=df_cleaned)
plt.title('Distribution of Clients Based on Encoded Attributes')
plt.xlabel('Attribute')
plt.ylabel('Number of Clients')
plt.show()
```
