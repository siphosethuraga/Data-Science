```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
import pandas as pd
import numpy as np
```

```python

ds = pd.read_excel(r'C:\Users\Devaire User\Desktop\Access python\Dots Potential Customer Survey Data.xlsx')
```

```python
ds.head()
```

```python
# Update the list of columns to drop
columns_to_drop = ['Unique ID', 'Country', 'Annual Family Income ($)', 'Time spent playing indoor sports', 
                   'Time spent playing outdoor sports', 'IQ', 'Whether parents have specs', 
                   'Migrated within country', 'Migrated overseas', 'Education Status', 
                   'Has Gym Subscription', 'Has OTT subscription', 'Number of friends', 
                   'Likes spicy food', 'Likes desserts', 'Wants to change career', 
                   'Has debt', 'Has kids', 'Drinks alcohol', 'Smoker']

# Drop the specified columns from the DataFrame
ds = ds.drop(columns=columns_to_drop)

# Display the first few rows of the modified DataFrame
print(ds.head())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop non-numeric columns
numeric_ds = ds.select_dtypes(include=['number'])

# 1. Heatmap
heatmap_data = numeric_ds.corr()  # Correlation matrix for numeric columns
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# 2. Line graph
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='Time spent watching videos/TV', data=ds)
plt.title('Age vs Time Spent Watching Videos/TV')
plt.xlabel('Age')
plt.ylabel('Time spent watching videos/TV')
plt.show()

# 3. Bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Sleeping hours', data=ds)
plt.title('Average Sleeping Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Sleeping hours')
plt.show()

# 4. Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='English speaker', y='Age', data=ds)
plt.title('Age Distribution by English Speaker')
plt.xlabel('English speaker')
plt.ylabel('Age')
plt.show()

# 5. Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=ds, x='Time spent watching videos/TV', bins=20, kde=True)
plt.title('Distribution of Time Spent Watching Videos/TV')
plt.xlabel('Time spent watching videos/TV')
plt.ylabel('Frequency')
plt.show()

# 6. Pie chart
plt.figure(figsize=(8, 8))
ds['Has Diabetes'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Percentage of Customers with Diabetes')
plt.ylabel('')
plt.show()

# 7. Scatter plot (without 'Wear Specs (Target Variable)')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Sleeping hours', data=ds)
plt.title('Age vs Sleeping Hours')
plt.xlabel('Age')
plt.ylabel('Sleeping hours')
plt.show()
```
